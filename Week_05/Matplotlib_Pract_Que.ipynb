# Practice Questions


### 1. import matplotlib.pyplot and create a basic line plot showing fare values of the first 10 passenger

```python
df1 = pd.read_csv("train.csv")
df1.head(10)
```

```python
df1.info()
```

```python

plt.plot(df1['Fare'].head(10))

plt.title('Fare of First 10 Titanic Passengers')
plt.xlabel('Passenger Index')
plt.ylabel('Fare')

plt.show()

```

### 2. create a bar chart showing the no of passenger in each class(Pclass).

```python

pclass_counts = df1['Pclass'].value_counts().sort_index()

plt.bar(pclass_counts.index, pclass_counts.values,color='red',edgecolor='black')

plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.grid()

plt.show()
```

### 3. plot a  histogram of  'age' column to visualize the distrubution of ages  

```python

plt.figure(figsize=(16, 9))
plt.hist(df1['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')

plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(color='skyblue')
plt.show()
```

### 4. create a scatter plot comparing 'age' and 'fare'. add label and a title to the plot 

```python
plt.scatter(df1.Age,df1.Fare,marker='o',color='orange');
plt.xlabel('age')
plt.ylabel('fare')
plt.title('age and fare distribution')
plt.grid()
```

## 5. Create a pie chart showing the percentage of survivors vs non-survivors.

```python
import pandas as pd
import matplotlib.pyplot as plt


survival_counts = df1['Survived'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%'])
plt.title('Survivors vs Non-Survivors')
plt.show()
```

## 7. Create two subplots side-by-side: - Left: Bar chart of survival count by gender - Right: Pie chart of 'Embarked' values

```python
gender_counts = df1['Sex'].value_counts()

gender_counts.plot(kind='bar')

plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Gender')
plt.show()
```

```python
embarked_counts = df1['Embarked'].value_counts()

plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Passenger Distribution by Embarkation Port')
plt.axis('equal')  
plt.show()
```

```python

plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
gender_counts = df1['Sex'].value_counts()

gender_counts.plot(kind='bar')

plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Gender')

plt.subplot(1,2,2)
embarked_counts = df1['Embarked'].value_counts()


plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Passenger Distribution by Embarkation Port')
plt.axis('equal')  
plt.show()

```

## 11. Create a stacked bar plot showing the number of survivors and non-survivors in each passenger class.

```python
import pandas as pd
import matplotlib.pyplot as plt

survival_counts = df1.groupby(['Pclass', 'Survived']).size().unstack()

survival_counts.columns = ['Did Not Survive', 'Survived']

survival_counts.plot(kind='bar', stacked=True, color=['pink', 'black'])


plt.title('Survivors and Non-Survivors by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Survival Status')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()

```
