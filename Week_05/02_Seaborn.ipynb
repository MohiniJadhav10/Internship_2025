```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

# histogram

```python
sns.histplot(tips['total_bill'], kde = False, bins = 15)
plt.title('total bill distribution');
```

# kde plot

```python
sns.kdeplot(tips['total_bill'], fill = True)
plt.title('KDE bill distribution');
```

# boxplot

```python
sns.boxplot(tips['tip']);
```

```python
# Find the upper and lower bound and apply boolean filtering

q3 = np.percentile(tips['tip'],75)
q1 = np.percentile(tips['tip'],25)
q3, q1
```

```python
iqr = q3 - q1
```

```python
iqr
```

```python
ub = q3 + iqr *1.5
lb = q1 - iqr *1.5
ub, lb
```

```python
x= tips[(tips['tip'] > lb) & (tips['tip'] > ub)]   # remove the outliers
```

```python
sns.boxplot(x['tip']);
```

# violin plot -> kde + boxplot

```python
sns.violinplot(tips);
```

# stripplot -> jittered scatter plot

```python
sns.stripplot(x = 'day', y = 'total_bill', data = tips, jitter= True);
```

# swarmplot

```python
sns.swarmplot(x = 'day', y = 'total_bill', data = tips);
```

# scatter plot

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = tips,
                hue = 'day');
```

# Regplot

```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips);
```
