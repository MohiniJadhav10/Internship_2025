```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2, 2, 1)
plt.title("line plot")
plt.xlabel("x values")
plt.ylabel("y values")

plt.plot(x, y, color = 'magenta', marker = '*', mfc = 'black', ls = '--');
plt.grid()
plt.subplot(2, 2, 2)
plt.title("line plot")
plt.xlabel("x values")
plt.ylabel("z values")

plt.plot(x, z, color = 'blue', marker = '*', mfc = 'red', ls = '--');
plt.grid()
```

```python

# Data
courses = ['python', 'dart', 'php', 'ruby', 'c++']
students = [56, 40, 30, 41, 65]
plt.figure(figsize=(12, 9))

# 1. Line Plot
plt.subplot(2, 2, 1)
plt.plot(courses, students, marker='o', color='blue')
plt.title("Line Plot")
plt.xlabel("Courses")
plt.ylabel("Students")

# 2. Bar Plot
plt.subplot(2, 2, 2)
plt.bar(courses, students, color='green')
plt.title("Bar Plot")
plt.xlabel("Courses")
plt.ylabel("Students")

# 3. Scatter Plot
plt.subplot(2, 2, 3)
plt.scatter(courses, students, color='red')
plt.title("Scatter Plot")
plt.xlabel("Courses")
plt.ylabel("Students")

# 4. Pie Chart
plt.subplot(2, 2, 4)
plt.pie(students, labels=courses, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart")

plt.savefig('my_plot.png')
```
