```python
x = df[[ 'Age', 'EstimatedSalary']]
y = df['Purchased']
```

```python
plt.title('age vs salary')
plt.xlabel('age')
plt.ylabel('salary')

plt.scatter(df['Age'], df['EstimatedSalary']);
```

```python
# cross validation
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y ,
                                                random_state = 0,
                                                test_size= 0.25)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# preprocessing
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df_scaled = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
df_scaled
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_scaled, y_train)
```

```python
# evaluation
```

```python
# predict the test
```

```python
y_pred = log.predict(x_test)
```

```python
y_pred
```

```python
resdf = pd.DataFrame({
    'Actual': y_test, 
    'Predicted': y_pred
})
```

```python
resdf
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, classification_report, accuracy_score
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred);
```

```python
print(classification_report(y_test, y_pred))
```

```python
recall = 67 / 68 # recall
# out of total 0 in the data how many model predicted
```

```python
precision = 67 / 77 # precision
# out o 0 predicted by model how many are correct
```

```python
(2 * recall * precision)/ (recall + precision)   # f1 score
# harmonic mean of presision and recall
```

```python
# actaul accuracy 
accuracy_score(y_test, y_pred)
```

```python
# rou_auc
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
y_pred_prob = log.predict_proba(x_test)[:,1]
```

```python
y_pred_prob
```

```python
auc = roc_auc_score(y_test, y_pred_prob)
```

```python
roc_curve(y_test,y_pred_prob)
```

```python
fpr, tpr, threshold = roc_curve(y_test,y_pred_prob)
```

```python
plt.title("AUC curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.plot(fpr, tpr, color = 'magenta', label = 'auc')
plt.plot([0,1], [0,1], label = 'random model', ls = '--')
plt.grid()
plt.legend();
```
